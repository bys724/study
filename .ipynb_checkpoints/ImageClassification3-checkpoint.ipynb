```python
import pathlib
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos', origin=dataset_url, untar=True)
data_dir = pathlib.Path(data_dir)


tmp = list(data_dir.glob('*/'))
classes = []
for i in tmp:
    if os.path.isdir(i):
        classes.append(os.path.basename(i))
print(classes)

Images = []
for class_ in classes:
    list_ = list(data_dir.glob(class_ + '/*'))
    Images.append(list_)
    

numBatches = 100
numImagesOfClassesForBatch = 4
width = 180
height = 180

inputs_train = np.empty([1, len(classes) * numImagesOfClassesForBatch, width, height, 3])
outputs_train = np.empty([1, len(classes) * numImagesOfClassesForBatch])
for i in range(int(numBatches * 0.7)):
    
    buffer = np.empty([1, width, height, 3])
    
    for class_ in range(len(classes)):
        numImgs = len(Images[class_])
        choice_idx = np.random.choice(numImgs, numImagesOfClassesForBatch, replace=False)
        choice = [Images[class_][j] for j in choice_idx]
        
        for path in choice:
            with PIL.Image.open(path) as img:
                im = resizeimage.resize_cover(img, [width, height])
                im = np.expand_dims(np.array(im), axis=0)
                buffer = np.concatenate((buffer, im), axis=0)
                
                
    buffer = np.delete(buffer, 0, axis=0)
    buffer = np.expand_dims(buffer, axis=0)
    
    output = np.empty([1,])
    for class_ in range(len(classes)):
        output = np.concatenate((output, np.ones(shape=(numImagesOfClassesForBatch)) * class_), axis=0)
    output = np.delete(output, 0, axis=0)
    output = np.expand_dims(output, axis=0)
    
    inputs_train = np.concatenate((inputs_train, buffer), axis=0)
    outputs_train = np.concatenate((outputs_train, output), axis=0)
    
    
inputs_train = np.delete(inputs_train, 0, axis=0)
outputs_train = np.delete(outputs_train, 0, axis=0)

dataset_train = tf.data.Dataset.from_tensor_slices((inputs_train, outputs_train))


inputs_valid = np.empty([1, len(classes) * numImagesOfClassesForBatch, width, height, 3])
outputs_valid = np.empty([1, len(classes) * numImagesOfClassesForBatch])
for i in range(int(numBatches * 0.3)):
    
    buffer = np.empty([1, width, height, 3])
    
    for class_ in range(len(classes)):
        numImgs = len(Images[class_])
        choice_idx = np.random.choice(numImgs, numImagesOfClassesForBatch, replace=False)
        choice = [Images[class_][j] for j in choice_idx]
        
        for path in choice:
            with PIL.Image.open(path) as img:
                im = resizeimage.resize_cover(img, [width, height])
                im = np.expand_dims(np.array(im), axis=0)
                buffer = np.concatenate((buffer, im), axis=0)
                
                
    buffer = np.delete(buffer, 0, axis=0)
    buffer = np.expand_dims(buffer, axis=0)
    
    output = np.empty([1,])
    for class_ in range(len(classes)):
        output = np.concatenate((output, np.ones(shape=(numImagesOfClassesForBatch)) * class_), axis=0)
    output = np.delete(output, 0, axis=0)
    output = np.expand_dims(output, axis=0)
    
    inputs_valid = np.concatenate((inputs_valid, buffer), axis=0)
    outputs_valid = np.concatenate((outputs_valid, output), axis=0)
    
    
inputs_valid = np.delete(inputs_valid, 0, axis=0)
outputs_valid = np.delete(outputs_valid, 0, axis=0)

dataset_valid = tf.data.Dataset.from_tensor_slices((inputs_valid, outputs_valid))
```

```python
def drawFeatureMap(folderName, epoch, layers, idxBatch):
    inputs = inputs_train[idxBatch,:,:,:,:]
    idx = tf.expand_dims(tf.range(len(classes)) * numImagesOfClassesForBatch, axis=-1)
    inputs_ = tf.gather_nd(inputs, idx)
    
    #print(inputs_.shape)
    
    numLayers = len(layers)
    layer_outputs = []
    
    for idxLayer in range(numLayers):
        layer_output = layers[idxLayer](inputs_)
        #print(layer_output.shape)
        
        outputs_abs = tf.abs(layer_output)
        maxValues = tf.reduce_max(outputs_abs, axis=-1)
        
        processed_layer_output = tf.where(tf.expand_dims(tf.greater(maxValues, 0.5), axis=-1), layer_output, 0)
        
        layer_outputs.append(processed_layer_output.numpy())
        
        inputs_ = layer_output
    
    inputs_ = tf.gather_nd(inputs, idx)
        
    for idx_layer, outputs in enumerate(layer_outputs):
        
        path_layer = folderName + '/layer ' + str(idx_layer)
        if not os.path.isdir(path_layer):
            os.mkdir(path_layer)
            
        for class_ in range(5):
            path_class = path_layer + '/class ' + str(class_)
            if not os.path.isdir(path_class):
                os.mkdir(path_class)
            
            output_ = outputs[class_, :,:,:]
            numFeatures = output_.shape[-1]
            
            width_ratios = [1] * ((numFeatures + 3) // 6 + 1)
            fig, axs = plt.subplots(6, (numFeatures + 3) // 6 + 1, gridspec_kw={'width_ratios': width_ratios}, figsize=(18, 15))

            idx_plot = 0
            for color in range(3):
                axs[color, 0].imshow(inputs_[class_,:,:,color], cmap='gray', vmin=0, vmax=255)
                idx_plot +=1
                
            for feature in range(numFeatures):
                featureMap = output_[:,:,feature]
                ax = axs[idx_plot%6, int(idx_plot // 6)]
                pcm = ax.pcolormesh(featureMap, cmap='RdBu_r')
                idx_plot +=1
                
            fig.colorbar(pcm, ax=axs[:, (numFeatures + 3) // 6], location='right')

            for i in range(6 - idx_plot%6):
                axs[-(i+1), -1].axis('off')
                
            plt.savefig(path_class + '/' + str(epoch) + '.jpg')
            plt.close(fig)
    
#drawFeatureMap(folderName = './savedImgs/test' , epoch = 0, model = model, idxBatch = 0)
```

```python
class customLoss(tf.keras.losses.Loss):
    
    def __init__(self, name="distanceLoss", ratio = 4):
        super().__init__(name=name)
        self.ratio = ratio
        
    def __call__(self, outputs, labels):
        
        shape = outputs.shape.as_list()[1:]
        shape.insert(0, 4)
        shape.insert(0, 5)
        outputs_separated = tf.reshape(outputs, shape = shape)
        outputs_separated_ = tf.reshape(outputs_separated, shape = (5, 4, -1, shape[-1]))
        
        
        outputs_abs = tf.abs(outputs_separated_)
        maxValues = tf.reduce_max(outputs_abs, axis=-1)
        
        
        deactivatedProcessed = tf.where(tf.expand_dims(tf.greater(maxValues, 0.5), axis=-1), outputs_separated_, 0)
        
        outputs_positive = tf.where(deactivatedProcessed > 0, deactivatedProcessed, 0)
        outputs_negative = tf.where(deactivatedProcessed < 0, deactivatedProcessed, 0)
        
        sumOfPositive = tf.reduce_sum(outputs_positive, axis=-2)
        sumOfNegative = tf.reduce_sum(outputs_positive, axis=-2)
        
        outputs_ratio =  sumOfPositive / (sumOfPositive + sumOfNegative)
        
        stdForClasses = tf.math.reduce_std(outputs_ratio, axis=1) # shape (numClass, numFeature)
        tmp1 = tf.square(stdForClasses)
        tmp2 = tf.reduce_sum(tmp1, axis=-1)
        tmp3 = tf.sqrt(tmp2 + kb.epsilon())
        L2 = tf.math.reduce_sum(tmp3)
        
        #L2 = tf.math.reduce_sum(stdForClasses)
        
        meanForClasses = tf.math.reduce_mean(outputs_ratio, axis=1) # shape (numClass, numFeature)
        stdForMeans = tf.math.reduce_std(meanForClasses, axis=0) # shape (numFeature)
        
        tmp4 = tf.square(stdForMeans)
        tmp5 = tf.reduce_sum(tmp4, axis=-1)
        tmp6 = tf.sqrt(tmp5 + kb.epsilon())
        
        
        L1 = tf.math.exp(-tmp6)
        
        Loss = L1 + L2 * self.ratio
        
        return Loss, L1, L2
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001)

L1_train = tf.keras.metrics.Mean()
L2_train = tf.keras.metrics.Mean()

@tf.function
def train_step(images, labels, ratio):
    
    loss = customLoss(name="distanceLoss", ratio = ratio)
    
    with tf.GradientTape() as tape:
        #tf.print(images)
        outputs = layer0(images)
        Loss, L1, L2 = loss(outputs, labels)
        
    gradients = tape.gradient(Loss, layer0.trainable_weights)
    optimizer.apply_gradients(zip(gradients, layer0.trainable_weights))
    
    L1_train(L1)
    L2_train(L2)
    
L1_valid = tf.keras.metrics.Mean()
L2_valid = tf.keras.metrics.Mean()
@tf.function
def test_step(images, labels):
    outputs = layer0(images)
    Loss, L1, L2 = loss(outputs, labels)
    
    L1_valid(L1)
    L2_valid(L2)
```
